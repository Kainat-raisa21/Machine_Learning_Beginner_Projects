```python
x = df[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]
y = df["Yearly Amount Spent"]

x1, x2, y1, y2 = train_test_split(x, y, test_size=0.33, random_state=101)

model = LinearRegression()
model.fit(x1, y1)

p = model.predict(x2)
p
```

```python
sns.relplot(x=y2, y=p)
```

```python
sns.displot(y2-p, color="seagreen")
plt.xlabel("actual yearly amounts spent ($)")
plt.ylabel("predicted amounts by model ($)")
```

```python
slope = model.coef_[0]
intercept = model.intercept_
```

```python
mae = metrics.mean_absolute_error(y2, p)
mae
```

```python
mse = metrics.mean_squared_error(y2, p)
mse
```

```python
rmse = np.sqrt(metrics.mean_squared_error(y2, p))
```
